```python
from ipywidgets import interact
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.0,1.0,0.01)

def plotsin(f):
    plt.plot(t,np.sin(2*np.pi*t*f))
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

interact(plotsin, f=(1,10,0.1))
```
